## Aterando nome das colunas

Para facilitar a escrita do código irei substituir o estilo de escrita das colunas para snake_case.

```python
columns = list(avocado_df.columns)

new_columns = [i.replace(" ", '_') for i in columns]

avocado_df.columns = new_columns

#alterando no as colunas 4046, 4225 e 4770 
avocado_df = avocado_df.rename(columns={"4046":"Small_Hass", "4225":"Large_Hass", "4770":"XLarge_Hass"})

avocado_df.head()
```

## Alterando coluna region

```python
region = list(avocado_df.region.unique())
region
```

```python
avocado_df.region.value_counts()
```

Existe um erro no preenchimento dos valores da coluna region. Os valores das regiões deveriam somar 504 porém, para os valores que tem espaço  ou barra em sua estrutura (por exemplo Los Angeles ou Baltimore/Washington), foram dividido em dois grupos um com o nome escrito com espaço/barra com um total de 402 valores, e ou com o nome concatenado (exemplo: LosAngeles e BaltimoreWashington) com um total de 104 valores.

```python
def Alteralinha(col):
    linha = col
    linha = col.replace(' ','')
    linha = linha.replace('.', '')
    linha = linha.replace('/', '')
    return linha
```

```python
avocado_df.region = avocado_df.region.apply(lambda x: Alteralinha(x))
avocado_df.region.value_counts()
```

```python
avocado_df.region.unique()
```

Irei remover os dados dos estados para evitar a multicolineariedade.

```python
regions_to_remove = ['California', 'GreatLakes', 'Midsouth', 'NewYork', 'Northeast', 'SouthCarolina', 'Plains', 'SouthCentral', 'Southeast', 'TotalUS', 'West']
avocado_df = avocado_df[~avocado_df.region.isin(regions_to_remove)]
avocado_df
```
